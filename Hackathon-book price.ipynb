```python
for i,ed in enumerate(train['Edition_part2']):
    if len(ed.split(','))>=3:
        #print(ed.split(','))
        train.drop(i,inplace=True)
```

```python
for i,ed in enumerate(train['Edition_part2']):
    if len(str(ed).split(','))>=2:
        train.loc[i,'EditionType']=ed.split(',')[0]
        train.loc[i,'BookDate']=ed.split(',')[1]
    else:
        train.loc[i,'BookDate']=str(ed).split(',')[0]
        train.loc[i,'EditionType']='Regular'
```
